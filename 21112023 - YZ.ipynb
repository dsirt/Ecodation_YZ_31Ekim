```python
#DF Birleştirme (Merging)
```

```python
daf1 = pd.DataFrame({'id' : ['1', '2', '3', '4', '5'], 'Name' : ['Dogu', 'Bati', 'Kuzey', 'Guney', 'Poyraz']})
daf1
```

```python
daf2 = pd.DataFrame({'id' : ['1', '2', '6', '7', '8'], 'Score' : [40, 60, 80, 90, 70]})
daf2
```

```python
#Inner Join
pd.merge(daf1, daf2, on = 'id', how = 'inner')
```

```python
#Full Outer Join
pd.merge(daf1, daf2, on = 'id', how = 'outer')
```

```python
#Left Outer Join
pd.merge(daf1, daf2, on = 'id', how = 'left')
```

```python
#Right Outer Join
pd.merge(daf1, daf2, on = 'id', how = 'right')
```

```python
path = r'C:\Users\ITU\mydata\COVID-19'
filenames = glob.glob(path + "/*.csv")
covid = pd.DataFrame()
for f in filenames:
    df = pd.read_csv(f)
    covid = covid.append(df, ignore_index = True, sort = True)
```
